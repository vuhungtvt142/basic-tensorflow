```
import tensorflow as tf 
print(tf.__version__)
```

```
dataset_mnist = tf.keras.datasets.fashion_mnist
(train_dataset, train_label), (test_dataset, test_label)= dataset_mnist.load_data()
```

```
train_dataset =  train_dataset/255.0
test_dataset =  test_dataset/255.0
```

```
model =  tf.keras.Sequential([
                              tf.keras.layers.Flatten(),
                              tf.keras.layers.Dense(512,activation=tf.nn.relu),
                              tf.keras.layers.Dense(10,activation=tf.nn.softmax)
])

model.compile(optimizer=tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics='accuracy')

class Callback(tf.keras.callbacks.Callback):
  def on_epoch_stop(self, epoch, logs={}):
    if(logs.get('loss')<0.15):
      print("\nReached 60% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = Callback()
model.fit(train_dataset,train_label,epochs=5, callbacks=[callbacks])
```

```
model.evaluate(test_dataset, test_label)
classifications = model.predict(test_dataset)

print(classifications[0])
print(test_label[0])
```
