```
import tensorflow_datasets as tfds
import tensorflow as tf

dataset_name = 'cats_vs_dogs'  

dataset, info =  tfds.load(name=dataset_name , split=tfds.Split.TRAIN, with_info=True)
# dataset1, info =  tfds.load(name=dataset_name , split=tfds.Split.TEST, with_info=True)
```

```
def preprocess(features):
    # YOUR CODE HERE
    image = tf.cast(features['image'], tf.float32),
    label =  tf.one_hot(features['label'], 2)
    
    return image, label
```

```
from tensorflow.keras.optimizers import RMSprop
def solution_model():
    train_dataset = dataset.map(preprocess).batch(32)
    # val_dataset = dataset1.map(preprocess).batch(32)
    # YOUR CODE TO TRAIN A MODEL HERE
    # Create network model 
    # model = tf.keras.models.Sequential([
    #         tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(300, 300, 3)),
    #         tf.keras.layers.MaxPooling2D(2,2),

    #         tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    #         tf.keras.layers.MaxPooling2D(2,2),

    #         tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    #         tf.keras.layers.MaxPooling2D(2,2),

    #          # Flatten the results to feed into a DNN
    #         tf.keras.layers.Flatten(), 
    #          # 512 neuron hidden layer
    #         tf.keras.layers.Dense(512, activation='relu'), 
    #         tf.keras.layers.Dropout(0.4),
    #         tf.keras.layers.Dense(2, activation='softmax')
    #     ])
    # Create model network
    model = tf.keras.Sequential([
      tf.keras.layers.Flatten(input_shape=(300,300,3)),
      tf.keras.layers.Dense(64, activation='relu'),
      tf.keras.layers.Dense(32, activation='relu'),
      # tf.keras.layers.Dense(64, activation='relu'),
      # tf.keras.layers.Dense(32, activation='relu'),
      # tf.keras.layers.Dropout(0.1),
      tf.keras.layers.Dense(2, activation='softmax'),
    ])
    # Loss function and optimization 
    model.compile(optimizer = RMSprop(lr=0.0001), 
              loss = 'categorical_crossentropy', 
              metrics = ['accuracy'])
    # Training model 
    model.fit(train_dataset, epochs=10)

    # model.summary()
    return model
```
